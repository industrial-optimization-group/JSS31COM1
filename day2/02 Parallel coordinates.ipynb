```python
u = np.linspace(0,1,100)
v1 = u + 10000000
v2 = -u
v3 = (0.25 - (u-0.5) ** 2) ** 0.5
rand = np.random.rand(100)
# v4 = Add other interesting patterns

data = pd.DataFrame(
    np.vstack((u, v1, v2, v3, rand)).T,
    columns = ["input", "proportional", "inverse", "circle", "random"],
)
```

```python
two_columns = ["input", "random"]
ex.scatter(data, x=two_columns[0], y=two_columns[1])
```
